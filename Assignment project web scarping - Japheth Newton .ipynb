```python
           #To perform web scraping in Python, you can use libraries such as BeautifulSoup,Request and pandas

#libries are imported
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Retrieve the HTML: Use a library like requests to send an HTTP request to the website and retrieve the HTML content of the page
url = 'http://quotes.toscrape.com/page/1/'
response = requests.get(url)
```

```python
r = response.content
```

```python
#Parse the HTML: Use the web scraping library (e.g., BeautifulSoup) to parse the HTML content and create a parse tree that represents the structure of the web page
soup = BeautifulSoup(r, 'html.parser')
```

```python
#Find Elements: Use the library's methods or functions to find specific HTML elements that contain the data you want to extract
El= soup.find("div")
```

```python

quotes = El.find_all("div",class_ ="quote" )
```

```python
#Extract Data: Once you've located the desired elements, use the library's methods to extract the relevant data. 
#This could involve accessing attributes, retrieving text and content.
```

```python
for div in quotes:
    quote = div.find("span").text[1:-1]
    print(quote)
```

```python
for div in quotes:
    author = div.find("small").text
    print(author)
```

```python
for div in quotes:
    tags = div.find("meta")
    tags_1 = tags.attrs["content"]
    print(tags_1)
```

```python
for div in quotes:
    top_10_tags = div.find("a" , class_ = "tag")
    top_10 = top_10_tags.attrs["href"][5:-8]
    print(top_10)
    
```
